# Import Libraries

```python
# Import libraries
import pickle
import pandas as pd
```

# Load Files

```python
# Set max columns for readibility
pd.set_option('display.max_columns', None)

# Load Files
with open('deployment/rf_randomcv_best.pkl', 'rb') as file_1:
  model = pickle.load(file_1)
```

# Inferencing

```python
# Load data
data_inf = pd.read_csv('test.csv', nrows=5)
data_inf.drop(columns=['Loan Status'], inplace=True)
```

```python
# Remove whitespace from column name
for col in data_inf.columns:
    data_inf.rename(columns={col: col.replace(' ', '')}, inplace=True)

data_inf.rename(columns={'EmploymentDuration': 'HomeOwnership',
                   'HomeOwnership': 'Income'}, inplace=True)

data_inf
```

```python
#Predict using linreg
y_pred_inf = model.predict(data_inf)
data_inf['Loan Status'] = y_pred_inf
```

Berdasarkan hasil prediksi dari model yang diterapkan pada data inference, dapat disimpulkan sebagai berikut:

Data 1, 2, dan 3: Model memprediksi bahwa individu dalam data ini non-defaulter (tidak gagal bayar). Hal ini menunjukkan bahwa fitur-fitur yang ada pada data pertama sesuai dengan pola atau karakteristik yang diidentifikasi oleh model sebagai indikasi positif untuk tidak mengalami gagal bayar.

Data 4 dan 5: Sebaliknya, model memprediksi bahwa individu dalam data ini termasuk ke dalam kelas defaulter (gagal bayar). Ini mengindikasikan bahwa fitur-fitur pada individu ini tidak memenuhi kriteria yang dianggap mengarah pada keputusan untuk membeli, atau mungkin menunjukkan karakteristik yang lebih mirip dengan konsumen yang tidak melakukan pembelian. Individu yang mengalami gagal bayar cenderung memiliki jumlah pinjaman yang lebih besar dibandingkan pinjaman yang tidak bermasalah. Ini mengindikasikan risiko lebih tinggi terkait pinjaman dengan jumlah besar.
